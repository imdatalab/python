### 문자열 조작, split과 join
- 문자열을 공백, "/", "," 등을 기준으로 나누거나 합친다
- 나누어진 항목은 리스트(list)에 저장된다 (리스트는 여러 데이터를 묶은 것)

```python
## 문자열을 ','나 공백(스페이스) 등으로 나누기

name1 = '김연아,손흥민,이선희,유재석'

# ','로 나누기 split 함수를 이용, z는 [] 리스트를 의미함
z = name1.split(',')
print(z)

# 스페이스로 나누기 (디폴트)
name2 = '김연아 손흥민 이선희 유재석'
print(name2.split())  # 디폴트로 공백을 기준으로 나눈다
```

```python
## 단어 합치기

';'.join(z)
```

```python
#join 함수 변형 
'--'.join(z)
```

```python
## 날짜를 "/"를 기준으로 나누고 순서를 바꾸어서 합치기

date = '12/31/2023'
z = date.split('/')
print(z)
print(z[1])
new_date = "-".join((z[2], z[0], z[1]))
print(new_date)
new_date = "::".join((z[2], z[0], z[1]))
print(new_date)
```

```python
# 문자열 교체 함수 replace

x = "오늘 날씨는 "
print(x.replace("날씨는", "기온은"))
```

### 문자열에 포맷( format) 사용하기
- 문자열을 화면에 출력하는 방법(포맷)을 지정할 수 있다
- 세가지 방법이 있다.
 - "%" 사용
 - format() 사용
 - f"..." 사용 (가장 간단한 방법)

- %를 사용하는 방법

```python
## 문자열 내에 변수를 같이 출력하는 방법 (format)

x = 5
y = 90.5

print(f"틀린 갯수가 {x}개 이하이면 {y} 점입니다.")
print("틀린 갯수가 %d개 이하이면 %f 점입니다." %(x,y))
print("틀린 갯수가 {0}개 이하이면 {1} 점입니다.".format(x,y))
```

## 튜플 (tuple)
- 튜플도 리스트처럼 여러 항목을 담을 수 있는 변수
- 리스트와 달리 튜플의 내용은 '상수'로 취급되어 값을 바꿀 수 없다
- 읽기 전용 리스트인 셈이며, 처리 속도가 빠르다

- 튜플로 슬라이싱이 가능하다
- 튜플을 만들려면 소괄호( )를 사용하여 만들거나 ( )  없이 "," 로 구분하여 나열한다

```python
## 튜플 만들기, 복제, 합치기
# 튜플은 append 함수를 사용할 수 없음(why? 튜플은 상수이기에 데이터 추가 안됨)

tuple_a = (1, 2, 3)
tuple_b = 90,"good"
z = tuple_a*2 + tuple_b
print(z)
print(z[-2])
print(3 in z)
print(4 in z)
```

## 사전 (Dictionaries)
- 키-값 조합 형태의 데이터 묶음
- 간단한 데이터베이스처럼 사용된다
- 항목을 찾을 때 키를 사용한다
- 딕셔너리에서 저장된 순서는 의미가 없다
- 딕셔너리는 중괄호 {  } 로 만들거나 dict()로 만든다

```python
## 딕셔너리 생성, 값변경, 항목 추가하기

student = {'grade': 3,'age': 21,'name': "kim"  }
print(student)
print(student['grade'])
student['age'] = 28
student['address'] = 'Seoul' #항목 추가도 가능함
print(student)
```

```python
## 딕셔너리의 키, 값, 항목 전체 출력하기

print(student.keys())
print(student.values())
print(student.items()) 
```

```python
## 딕셔너리에 특정 키가 존재하는지 확인하기

print("age" in student)
print("sex" in student)
```

## if
- if 문 다음이 True이면 ":" 이후 블록을 실행한다
- 블록은 들여쓰기가 같은 영역으로 구분한다(들여쓰기를 잘 해줘야 함!!)
- else: 문을 사용할 수도 있다.
- if 뒤에는 불리언이 필요하다

```python
## 기본적인 if, elif, else 사용법

x = 90

if x > 80:
    print("축하합니다~ 합격입니다")
    print("다음주부터 출근하시면 됩니다.")
elif x > 60:
    print("2차 면접시험 대상자 입니다.")
    print("시험 날짜를 알려드리겠습니다.")
else:
    print("불합격입니다.")
```

```python
## for 문을 사용하여 채점 결과를 알려주는 코드

score = [50, 70, 90]
for x in score: # for문에서의 in은 score에 있는 점수를 하나씩 꺼내서 반복하라는 의미
    print('----------------------')
    if x > 80:
        print(f"점수 {x} 축하합니다~ 합격입니다")
        print("다음주부터 출근하시면 됩니다.")
    elif x > 60:
        print(f"점수 {x} 2차 면접시험 대상자 입니다.")
        print("시험 날짜를 알려드리겠습니다.")
    else:
        print(f"점수 {x} 불합격입니다.")
```

```python
## for 문을 사용하여 채점 결과를 알려주는 코드

score = [50, 70, 90]
for x in score: # for문에서의 in은 score에 있는 점수를 하나씩 꺼내서 반복하라는 의미
    print('----------------------')
    print(x, "점 입니다")
    if x > 80:
        print("축하합니다~ 합격입니다")
        print("다음주부터 출근하시면 됩니다.")
    elif x > 60:
        print("2차 면접시험 대상자 입니다.")
        print("시험 날짜를 알려드리겠습니다.")
    else:
        print("불합격입니다.")
```

### 구구단을 딕셔너리 구조의 데이터베이스로
- 결과를 딕셔너리로 만들고 입력을 주면 출력을 리턴하는 간단한 데이터베이스로 만들어보시오
- result[(7,9)] ==> 결과 63
- 힌트: 빈 딕션너리를 만들고 키 값으로 (2,1), (2,2),..등을 사용한다

```python
## 구구단을 딕셔너리로 만들고 사용하기

result = {}
for i in range(2,10):
    for j in range(1,10):
        result[(i, j)] = i*j
print(result)
        
# 구구단 사용하기
result[(7,9)] # 63
```
